Question 1 :
Define a function, which takes in two strings A and B. Reverses the first string A, adds it to B, and returns the final string.

```python
def reverse_and_concatenate(A, B):
    reversed_A = A[::-1]  # Reverse string A
    result = reversed_A + B  # Concatenate the reversed A with B
    return result


string_A = "Hello"
string_B = "World"
final_string = reverse_and_concatenate(string_A, string_B)
print(final_string)
```

Question 2 :
Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]

```python
def calculate_highest_paid_female(work_hours):
    highest_hours = 0
    highest_paid_female = None

    for name, hours, gender in work_hours:
        if gender == 'F' and hours > highest_hours:
            highest_hours = hours
            highest_paid_female = name

    if highest_paid_female:
        pay = highest_hours * 20
        print(f"The female worker that worked the most hours is {highest_paid_female} and she should be paid ${pay}.")
    else:
        print("No female workers found in the list.")


work_hours = [('Abby', 100, 'F'), ('Billy', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'), ('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]

calculate_highest_paid_female(work_hours)
```

i.e. each row is increasing and has repetive elements.

b) find dot product of the matrix with any matrix. (Figure out the size/ shape of the matrix)

```python
import numpy as np
pattern_matrix = np.array([[1, 1],
                            [2, 2]])
result_matrix = np.vstack([np.hstack([pattern_matrix, pattern_matrix]),
                           np.hstack([pattern_matrix, pattern_matrix])])
print(result_matrix)
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
if mat1.shape[1] == mat2.shape[0]:
    dot_product = np.dot(mat1, mat2)
    print("Dot Product Result:")
    print(dot_product)
else:
    print("Matrix dimensions are not compatible for dot product.")
```
